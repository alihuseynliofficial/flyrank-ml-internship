## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My chosen lane is Content Refresh / Opportunity Scoring.

The main ML task is scoring because the system will assign each content page a refresh opportunity score. A higher score means that the page should receive more attention and may need to be updated sooner.

The scores can also be used to rank pages, helping the content team decide which pages to review first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is whether a content page is showing a decline in performance.

Because the dataset does not contain the real result of refreshing each page, I will use performance decline as a proxy for refresh need. The proxy label can be created from observed historical performance, such as comparing recent impressions with impressions from an earlier period.

The model will predict the probability of decline, and this probability can contribute to the refresh opportunity score.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The main success metric is Precision@50.

Precision@50 measures how many of the 50 highest-scored pages are actually
showing a decline in performance. This metric matches the real action because
the content team has limited time and will review only the highest-priority
pages.

I would consider a Precision@50 of at least 0.70 good. This means that at least
35 of the top 50 recommended pages are correctly identified as declining.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule based only on trend percentage would not be enough because a large percentage decline does not always mean a large business impact.

For example, a page that falls from 13 impressions to 1 impression has a very large percentage decline, but it may not be as important as a page that falls from 13,828 impressions to 5,696 impressions.

Machine learning can combine several signals such as search volume, current impressions, average position, engagement rate, content age, and days since the last update. This allows the system to identify more valuable refresh opportunities instead of depending on only one rule.

The output supports the content team by ranking pages and helping them decide which content should be reviewed and refreshed first.
